## Import Dependencies


```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data collection & Processing

```python
# load the data from sklearn
breast_cancer_data=sklearn.datasets.load_breast_cancer()
```

```python
# loading the data to a data frame
df= pd.DataFrame(data=breast_cancer_data.data, columns=breast_cancer_data.feature_names)
```

```python
df.head()
```

```python
# adding the target table
label=df['label']= breast_cancer_data.target
```

```python
##Statistical measures about the data
df.describe()
```

##### 1 ---> Bengin
##### 0 --->Malignant

```python
sns.countplot(x='label', data=df)
sns.set_theme(rc={'figure.figsize':(5,5)})
```

```python
df.groupby('label').mean()
```

```python
## separating the features and the target
X= df.drop(columns='label', axis=1)
y= df['label']
```

## Splitting the data into training data and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

## Model Training

```python
LR=LogisticRegression()
```

```python
## training the model using training data
LR.fit(X_train, y_train)
```

## Model Evaluation

```python
## accuracy on training data
X_train_pred= LR.predict(X_train)
training_acc=accuracy_score(y_train, X_train_pred)
```

```python
print('Accuracy of training data = ', training_acc)
```

```python
## accuracy on test data
X_test_pred= LR.predict(X_test)
testing_acc=accuracy_score(y_test, X_test_pred)
```

```python
print('Accuracy of test data = ', testing_acc)
```
